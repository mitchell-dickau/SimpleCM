```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to create lists
def createList(n1, n2, int):
    return np.arange(n1, n2+1, int)
```

```python
# Create year var
year = createList(2020, 2100, 1)
```

```python
# Create CO2 pathway function
def CO2_pathway(timestep, interval):
    path = []
    for x in timestep:
        if x == 2020:
            pre_val= 40 
            path.append(pre_val)
            continue 
        pre_val += interval
        if pre_val < 0:
            pre_val = 0
        path.append(pre_val)
        if x == 2100:
            return path
```

```python
# Create 3 CO2 pathways 
CO2_up = CO2_pathway(year, 0.5)
CO2_down = CO2_pathway(year, -0.5)
CO2_fastdown = CO2_pathway(year, -1.3)
CO2_stable = CO2_pathway(year, 0)

pathways = {'year':year, 'up':CO2_up, 'down':CO2_down, 'fastdown':CO2_fastdown, 'stable':CO2_stable}
```

```python
# Plot emissions pathways 
plt.plot(pathways['year'], pathways['up'], color='red')
plt.plot(pathways['year'], pathways['stable'], color='orange')
plt.plot(pathways['year'], pathways['down'], color='gold')
plt.plot(pathways['year'], pathways['fastdown'], color='blue')

plt.title("CO2 emissions pathways")
plt.xlabel("Year")
plt.ylabel("GtCO2")

plt.grid(color = 'grey', linestyle = '--', linewidth = 0.25)

plt.show()
```

```python
# Calculate temperature change 
em = ['up','stable','down','fastdown']
for x in range(4):
    dummy = pathways[em[x]]
    dummy2 = [1.2]
    for n in range(1,len(dummy)):
        temp = sum(dummy[1:n+1])/3.664*1.65/1000+1.2
        dummy2.append(temp)
    pathways[em[x]]= (pathways[em[x]], dummy2)
```

```python
# Plot emissions pathways 
cols= ['red','orange','gold','blue']
for n in range(0,len(em)):
    plt.plot(pathways['year'], pathways[em[n]][1], color= cols[n])
    
plt.title("Temperature")
plt.xlabel("Year")
plt.ylabel("C")

plt.grid(color = 'grey', linestyle = '--', linewidth = 0.25)

plt.show()

# plt.plot(pathways['year'], pathways['up'][1], color='red'
```
